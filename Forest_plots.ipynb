```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import font_manager as fm

# Reverse y-order so first row appears at the top of the plot
y_pos = list(range(len(df)))
y_pos.reverse()

# Figure
fig_forest, ax_forest = plt.subplots(figsize=(5, 6))

# Hide top/left/right spines; keep bottom
for spine in ["top", "left", "right"]:
    ax_forest.spines[spine].set_visible(False)
ax_forest.spines["bottom"].set_visible(True)
ax_forest.spines["bottom"].set_linewidth(1.3)

# Reference line at HR = 1
ax_forest.axvline(1.0, color='gray', linestyle=':', linewidth=1.4)

# Plot hazard-ratios with CI bars
ax_forest.errorbar(
    df["HR"], y_pos,
    xerr=[df["HR"] - df["CI_low"], df["CI_high"] - df["HR"]],
    fmt='s', color='#355360', ecolor='black',
    capsize=0, markersize=12, elinewidth=0.9
)

# Log-scale settings
ax_forest.set_xscale('log')          # switch to log scale
ax_forest.set_xlim(0.25, 1.5)         # keep original numeric bounds

# Show only lower bound, HR = 1, and upper bound
ax_forest.set_xticks([0.25, 1, 1.5], minor=False)
ax_forest.set_xticklabels(['0.25', '1', '1.5'], fontsize=font_size)

# Remove automatic minor ticks that Matplotlib adds to log axes
ax_forest.xaxis.set_minor_locator(ticker.NullLocator())

# Make major tick marks longer/bolder (same as before)
ax_forest.tick_params(axis='x', length=6, width=1.3)

# Labels
ax_forest.set_ylim(-1, len(df) - 0.5 + 4.5)   # top padding only
ax_forest.set_yticks([])                      # hide y tick labels

# X-axis label centred under the HR = 1 tick
ax_forest.text(1, -2.2, "HR (95% CI)", ha='center', va='top', fontsize=font_size)

# Render
plt.tight_layout()
plt.show()
plt.close(fig_forest)
```
